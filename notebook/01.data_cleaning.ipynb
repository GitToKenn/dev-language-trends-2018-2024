## 2024

```python
df_2024 = pd.read_csv("survey_results_public_2024.csv")
```

```python
df_2024_raw = df_2024.copy()
```

```python
df_2024_raw.columns.values
```

```python
df_2024_raw.head(2)
```

```python
# will work on the "LanguageHaveWorkedWith"
```

```python
df_2024_proc = df_2024_raw[['ResponseId', 'LanguageHaveWorkedWith']].dropna().copy()
```

```python
df_2024_proc.head()
```

```python
df_2024_proc['ResponseId'].nunique()
```

```python
# seperating the langauge from each ids
```

```python
df_2024_lang = df_2024_proc[['ResponseId', 'LanguageHaveWorkedWith']].dropna().copy()
```

```python
#split by ';'
df_2024_lang['LanguageHaveWorkedWith'] = df_2024_lang['LanguageHaveWorkedWith'].str.split(';')

#explode the lists into one row per language
df_2024_lang = df_2024_lang.explode('LanguageHaveWorkedWith')
```

```python
#cleaning up the spacing around the langauge name , if there is
df_2024_lang['LanguageHaveWorkedWith'] = df_2024_lang['LanguageHaveWorkedWith'].str.strip()
```

```python
#add year
df_2024_lang['Year'] = 2024
```

```python
# Add a GlobalRespondentID to ensure unique identifiers across all survey years
df_2024_lang['GlobalRespondentID'] = (
    df_2024_lang['Year'].astype(str) + '_' + df_2024_lang['ResponseId'].astype(str)
)
```

```python
df_2024_lang.head(30)
```

```python
sorted(df_2024_lang['LanguageHaveWorkedWith'].unique())
```

```python
len(df_2024_lang["LanguageHaveWorkedWith"].unique())
```

```python
len(df_2024_lang)
```

```python
df_2018_raw.head()
```

```python
df_2018_raw.columns.values
```

```python
# same as before, Respondent , also with the LanguageWorkedWith, will change to suit thee other year
```

```python
df_2018_proc = df_2018_raw[['Respondent', 'LanguageWorkedWith']] \
    .rename(columns={'Respondent': 'ResponseId', 'LanguageWorkedWith': 'LanguageHaveWorkedWith'}) \
    .dropna() \
    .copy()
```

```python
df_2018_proc.head(3)
```

```python
df_2018_lang['LanguageHaveWorkedWith'] = df_2018_lang['LanguageHaveWorkedWith'].apply(
    lambda x: ['HTML/CSS' if lang.strip() in ['HTML', 'CSS'] else lang.strip() for lang in x]
)
```

```python
#explode the lists into one row per language
df_2018_lang = df_2018_lang.explode('LanguageHaveWorkedWith')
```

```python
#cleaning up the spacing around the langauge name , if there is
df_2018_lang['LanguageHaveWorkedWith'] = df_2018_lang['LanguageHaveWorkedWith'].str.strip()
```
